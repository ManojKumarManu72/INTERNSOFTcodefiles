```python
a= pd.Series(np.random.normal(size=5))
print(a)
a.plot()
```

```python
a.cumsum().plot(grid=False)
```

```python
a.cumsum().plot(grid=True)
```

```python
b= pd.DataFrame({'normal': np.random.normal(size=5), 'gamma': np.random.gamma(1, size=5), 'poisson': np.random.poisson(size=5)})
print(b)
b.cumsum(0).plot()
```

```python
b.cumsum(0).plot(subplots=False)
```

```python
b.cumsum(0).plot(subplots=True)
```

```python
b.cumsum(0).plot(secondary_y='normal')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(4, 12))
for i,var in enumerate(['normal','gamma','poisson']):
    b[var].cumsum(0).plot(ax=axes[i], title=var)
axes[0].set_ylabel('cumulative sum')
```
